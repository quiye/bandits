```python
import numpy as np
from numpy.random import *
import matplotlib.pyplot as plt

trials=2000
params=[[10,20],  #avg is 10
        [13,11],
        [15,10],
        [17,8]  #avg is 17
       ]
def slot(i):
    return normal(params[i][0],params[i][1])

plt.show()

avg=0
lis=[]
for i in range(1,trials):
    avg+=slot(randint(len(params)))
    lis.append(avg/i)
plt.plot(lis,label='random')
print("random ",avg/len(lis))

avg=0
lis=[]
eps=100
epslist=[0]*4
maxx=-111111
arg=0
for i in range(1,trials):
    if(i<eps*len(params)):
        idx=i%len(params)
        slt=slot(idx)
        avg+=slt
        epslist[idx]+=slt
        if(maxx<epslist[idx]):
            maxx=epslist[idx]
            arg=idx
    else:
        avg+=slot(arg)
    lis.append(avg/i)
plt.plot(lis,label='epsilon-greedy')
plt.legend()
print("eps-grd",avg/len(lis)," by slot No.",arg)

plt.title("comparison of Bandit-Algorithms")
plt.xlabel("number of trials")
plt.ylabel("average gain")
plt.show()
```
